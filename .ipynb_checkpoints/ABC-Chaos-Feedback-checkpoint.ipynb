```python
from scipy.integrate import odeint
import math 
import numpy as np
import astroabc
import GPy
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from IPython.display import display

%matplotlib inline
```

```python
def model_deriv(y, t, param):
    #Define parameters
    a,b,c=param

    #define states
    X,Y,Z=y


    #define derivatives
    dX_dt = -Y - Z
    dY_dt = X + a*Y
    dZ_dt = b + Z*(X-c)
    
    return dX_dt,dY_dt,dZ_dt
    
def model_sol(param):
    
    y0 = [4.,4.,0.1]
    time = np.linspace(0, 100, 1000)

    solution = odeint(model_deriv, y0, time, args=(param,))
    return np.array(solution)
```

```python
time = np.linspace(0, 100, 1000)

sol=model_sol([0.1,0.1,14.])
Y=np.ones((len(time),3))
for states in range(3):
    Y[:,states] = sol[:,states] + np.random.randn(len(time),)*(0.1*np.std(sol[:,states]))

plt.plot(Y[:,0],Y[:,1])
plt.plot(sol[:,0],sol[:,1])
Y[:,0].shape
```

```python
kernX = GPy.kern.Matern32(1) #+ GPy.kern.Bias(1)
kernY = GPy.kern.Matern32(1)
kernZ = GPy.kern.Matern32(1)
input_time=time[-200:].reshape(200,1)
modelX = GPy.models.GPRegression(input_time, Y[-200:,0].reshape(len(input_time),1), kernX)
modelY = GPy.models.GPRegression(input_time, Y[-200:,1].reshape(len(input_time),1), kernY)
modelZ = GPy.models.GPRegression(input_time, Y[-200:,2].reshape(len(input_time),1), kernZ)
modelZ.kern.variance=1.342424
```

```python
def simulation_rhs_f(trial):
    print(trial)
    stateLen = 200
    sample = 10
    modelX.Mat32.variance, modelX.Mat32.lengthscale, modelX.Gaussian_noise.variance = trial[-9:-6]
    modelY.Mat32.variance, modelY.Mat32.lengthscale, modelY.Gaussian_noise.variance = trial[-6:-3]
    modelZ.Mat32.variance, modelZ.Mat32.lengthscale, modelZ.Gaussian_noise.variance = trial[-3:]
    
    
    xbar_X=modelX.posterior_samples_f(input_time,size=10)
    dX_f=modelX.predictive_gradients(input_time)
    xbar_Y=modelY.posterior_samples_f(input_time,size=10)
    dY_f=modelY.predictive_gradients(input_time)
    xbar_Z=modelZ.posterior_samples_f(input_time,size=10)
    dZ_f=modelZ.predictive_gradients(input_time)

    
    
    xbar_X, xbar_Y, xbar_Z = xbar
    rhsf_X = np.ones((stateLen,sample))
    rhsf_Y = np.ones((stateLen,sample))
    rhsf_Z = np.ones((stateLen,sample))
    for t in range(stateLen):
        for j in range(sample):
            rhsf_X[t,j] = -xbar_Y[t,j] - xbar_Z[t,j]
            rhsf_Y[t,j] = xbar_X[t,j] + trial[0]*xbar_Y[t,j]
            rhsf_Z[t,j] = trial[1] + xbar_Z[t,j]*(xbar_X[t,j]-trial[2])
    
    if np.any(np.array([trial])<0.0):
        rhsf_X[0,:]=-1000.
        rhsf_Y[0,:]=-1000.
        rhsf_Z[0,:]=-1000.
       
        
    return {'State':[xbar_X.mean(axis=1),xbar_Y.mean(axis=1),xbar_Z.mean(axis=1)], \
            'RHS': [rhsf_X.mean(axis=1),rhsf_Y.mean(axis=1),rhsf_Z.mean(axis=1)], \
            'Velocity':[dX_f[0].reshape(len(input_time),),dY_f[0].reshape(len(input_time),), \
                        dZ_f[0].reshape(len(input_time),)]}

```

```python
def dist_metric(d,x):
    dist1 = 0.0
    dist2 = 0.0
    
    State = x['State']
    RHS = x['RHS']
    Velocity = x['Velocity']
    
    for states in range(3):
        #dist += np.dot((d[states]-x[states]),(d[states]-x[states]).T)
        
        if np.all(np.array(RHS[states][0]==-1000.0)):
            dist += np.inf
        else:
            dist1 += np.sum((Velocity[states]-RHS[states])**2)    
            dist2 += np.sum((d[:,states] - State[states])**2) 
    
    return dist1 + dist2
    
```
